# **Data Cleaning**

The process of Data cleaning starts as folllowing :





* Optimize dataset while removing the duplicates.
*  Finding the null values in dataset , either assign 0 values to that column or remove that columns if that column are not important for analysis 








```python
 # finding duplicates value in dataset
df1.duplicated().value_counts() # true specifies total number of duplicates values
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df1.duplicated())
```

```python
# to optimize our dataset we are going to deleate duplicates values
df1 = df1.drop_duplicates()
```

```python
# now , the shape has been changed here.
df1.shape
```

```python
# finding the null values in dataset
df1.isnull().sum()
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
sns.heatmap(df.isnull(), cbar=False,cmap='flare',yticklabels=False)
plt.xlabel("Name Of Columns", fontsize=18)
plt.title('Finding missing values in columns',fontsize=25)
plt.show()
```

In the process of finding null values we find 4 columns that have null values :

*  country
*  children
*  company
*   agent

---




here we replace NA values to 0 in   children ,company columns , mean value in agent and 



"other" in country columns

```python
df1['agent'].fillna(df1["agent"].mean(), inplace=True )
```

```python
df1['children'].fillna(0, inplace=True )
```

```python
df1['company'].fillna(0, inplace=True )
```

```python
df1['country'].fillna("other", inplace=True )
```

```python
df1.isnull().sum()
```

```python
len(df1[df1['adults']+df1['children']+df1['babies']==0])
df1.drop(df1[df1['adults']+df1['babies']+df1['children']==0].index,inplace=True)
```

```python
column_tobe_used = list(df1["hotel","meal","country","reserved_room_type","assigned_room_type","agent", "customer_type"])
```

```python
df1['total_people'] = df1['adults']+df1['children']+df1['babies']
df1['total_stay']=df1['stays_in_weekend_nights']+df1['stays_in_week_nights']
```

5) who are the  top 5 agents who booked  highest number of hotels for customer? 

```python
# using groupby function in agent columns to find 10 popular agents 
highest_booking_agent = df1.groupby(['agent'])['agent'].agg(['count']).reset_index().rename(columns={'count': 'highest_bookings'}).sort_values(by = 'highest_bookings')
# getting 10 popular agent nlargest
popular_agents_bookings= highest_booking_agent.nlargest(10, ['highest_bookings'])
popular_agents_bookings
```

5) Which agent has made the highest bookings? 

```python
# visualize popular agents and their bookings through Bar plot
sns.barplot(x=popular_agents_bookings['agent'],y=popular_agents_bookings['highest_bookings'],order=popular_agents_bookings['agent'])
plt.rcParams["figure.figsize"] = (12,10)
plt.title("Highest Booking made by agent ", fontsize = 20)
plt.xlabel("agent_number", fontsize= 20)
plt.ylabel(" 10_highest_bookings", fontsize= 20)
plt.show()
```

**Observation**

9.0 is the agent who has highest booking.
